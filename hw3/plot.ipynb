```python
import matplotlib.pyplot as plt

data = [
    ("PCA", [2, 4, 6, 10]),
    ("ICA", [2, 4, 6, 10]),
    ("RP", [2, 4, 6, 10]),
    ("t-SNE", [2, 3]),
]

accuracy = [
    [0.56, 0.68, 0.68, 0.74],
    [0.54, 0.67, 0.68, 0.73],
    [0.32, 0.63, 0.65, 0.65],
    [0.28, 0.43],
]

x = data[0][1]

fig, ax = plt.subplots()

for i, (label, values) in enumerate(data):
    ax.plot(data[i][1], accuracy[i], label=label)

ax.set_xlabel("Dimension")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy vs. Dimension for Different Algorithms")
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
algorithms = ['K-Means', 'K-Means', 'K-Means', 'K-Means', 'EM', 'EM', 'EM', 'EM']
clusters = [0, 1, 2, 3, 0, 1, 2, 3]
missclassified = [23, 45, 91, 41, 21, 31, 51, 17]

# Create a bar graph
plt.figure(figsize=(10, 6))
plt.bar(range(len(algorithms)), missclassified, color=['blue', 'blue', 'blue', 'blue', 'green', 'green', 'green', 'green'])

# Set the x-axis labels
x_labels = [f"{algorithm}\nCluster {cluster}" for algorithm, cluster in zip(algorithms, clusters)]
plt.xticks(range(len(algorithms)), x_labels, rotation=0)

# Set axis labels and title
plt.xlabel('Algorithm and Cluster')
plt.ylabel('Missclassified Numbers')
plt.title('Missclassified Numbers in Clusters')

# Show the bar graph
plt.show()
```
